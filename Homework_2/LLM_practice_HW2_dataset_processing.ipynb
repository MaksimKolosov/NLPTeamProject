```python
import pandas as pd
```

```python
df_true = pd.read_csv("True.csv")
df_true["target"] = "1"
df_false = pd.read_csv("Fake.csv")
df_false["target"] = "0"
```

```python
def process_text(text):
    """
    Функция для обработки текста:
    - Если первые три буквы заглавные, удаляем первые три слова
    - Иначе возвращаем текст как есть
    """
    words = str(text).split()

    # Проверяем, есть ли хотя бы 3 слова и первые три буквы заглавные
    if len(words) >= 3 and text[:3].isupper() and text[:3] != "AUS" and text[:2] != "US":
        # Возвращаем текст без первых трех слов
        return ' '.join(words[3:])
    else:
        # Возвращаем текст как есть
        return text
def prefinal_clean(text):
    """
    Функция для остаточной обработки текста:
    удаляем первые символы, если это ("Fla ", "- ", "-", " ")
    """
    text_str = str(text)
    prefixes = ["Fla ", "Fla./WASHINGTON", "(Reuters) ", "(Reuters)""- ", "-", " "]
    for i in range(4):
        for prefix in prefixes:
           if text_str.startswith(prefix):
                return text_str[len(prefix):]

    return text_str
```

```python
df_true['text'] = df_true['text'].apply(process_text)
df_true['text'] = df_true['text'].apply(prefinal_clean)
df_true['text'] = df_true['text'].apply(prefinal_clean)
```

Мы избавились от основных проблем текста, однако надо убедиться, что слова (Reuters) не осталось в тексте.

```python
df_true['text'] = df_true['text'].str.replace('\(Reuters\)\s?', '', regex=True)
```

```python
#Теперь удаляем множественные пробелы, которые могли образоваться в результате наших действий или опечаток автора
df_true['text'] = df_true['text'].str.replace(r'\s+', ' ', regex=True)
df_true['text'] = df_true['text'].apply(prefinal_clean)
```
